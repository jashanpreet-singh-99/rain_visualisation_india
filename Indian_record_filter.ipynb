```python
# convert the parent database into subparts based on Year

l = list(database.YEAR.unique())

path = os.getcwd()
print(path)
for year in l :
    data = database[database.YEAR == int(year)].copy()
    data.reset_index(drop=True, inplace= True)
    data.to_csv(path + '/year_record/record_' + str(year) + '.csv')
```

```python
# convert the sepreated row datset based on year into filtered datasets

year_list = list(database.YEAR.unique())

for year in year_list: 
    data = rfs.read_data_csv(year)
    data = rfs.detect_presence_of_and(data)
    data = rfs.remove_the_and(data)
    data = rfs.remove_unwanted_words(data)
    data = rfs.detect_presence_of_space(data)
    data = rfs.remove_the_space(data)
    data = rfs.group_state_and_get_means(data)
    data = rfs.fix_error_prone_states(data)
    rfs.save_filtered_data(data)
    print('.', end="")

print("Done Filtering.")
```
